```python
from rasterio.windows import Window
from mantle_utils.alg.raster import Raster
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
start_date = pd.to_datetime('2019-01-01')
end_date = pd.to_datetime('2021-12-31')
df08 = pd.read_pickle('./namelist/s2_08.pkl')
df08['start_datetime'] = pd.to_datetime(df08['start_datetime'])
sorted_datetimes = sorted(list(set(df08['start_datetime'].values)))
```

```python
size = [80,50]
window = Window(col_off=100, row_off=200, width=size[0], height=size[1])
ip = []
for start_datetime_np in sorted_datetimes:
    geotiff_path = df08.loc[df08['start_datetime'] == start_datetime_np]['raster_path'].values[0]
    partial_raster = Raster(geotiff_path, window=window)
    temp = partial_raster.array.reshape(size[1],size[0])
    ip.append(temp)
```

```python
print(type(ip))
len(ip)
ip = np.array(ip)
```

```python
four_three = ip[:,45,32]
four_three = four_three.astype('float')
four_three[four_three == 0] = 'nan'
print(four_three)
```

```python
two_one = ip[:,25,16]
two_one = two_one.astype('float')
two_one[two_one == 0] = 'nan'
two_one
```

```python
%matplotlib inline
df = pd. DataFrame({'dates':sorted_datetimes,'data':two_one}, columns=['dates','data'])
df.dropna(inplace=True)
df.set_index('dates',inplace=True)
df.index=pd.to_datetime(df.index)
plt.plot(df.index,df['data'],  linestyle='-',marker='.')
print(df.head())
```

```python
df = pd. DataFrame({'dates':sorted_datetimes,'data':four_three}, columns=['dates','data'])
df.dropna(inplace=True)
df.set_index('dates',inplace=True)
df.index=pd.to_datetime(df.index)
plt.plot(df.index,df['data'],  linestyle='-',marker='.')
```
