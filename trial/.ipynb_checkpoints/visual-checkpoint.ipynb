```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-poster')
df = pd.read_csv('b1.csv')
```

```python
df.describe()
```

```python
fig,(ax0,ax1,ax2) = plt.subplots(1,3, figsize=(25,10))
ax0 = df.plot.scatter(x='Month', y='Day Type', c='Equipment Electric Power [kWh]',colormap='viridis')
ax0.set_ylabel('Day type')
ax0.set_xlabel('Month')
ax0.legend()
ax1 = df.plot.scatter(x='Month', y='Hour', c='Equipment Electric Power [kWh]',colormap='viridis')
ax1.set_ylabel('Hour')
ax1.set_xlabel('Month')
ax1.legend()
ax2 = df.plot.scatter(x='Day Type', y='Hour', c='Equipment Electric Power [kWh]',colormap='viridis')
ax2.set_ylabel('Hour')
ax2.set_xlabel('Day Type')
ax2.legend()
plt.show()
plt.close()
```
