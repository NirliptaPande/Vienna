```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-poster')
df = pd.read_csv('b1.csv')
```

```python
df.describe()
```

```python
fig,(ax0,ax1,ax2) = plt.subplots(1,3, figsize=(25,10))
ax0.plot(df['Month'],df['Day Type'], c='Equipment Electric Power [kWh]')
ax0.set_ylabel('Day type')
ax0.set_xlabel('Month')
ax1.plot(df['Month'],df['Hour'], c='Equipment Electric Power [kWh]')
ax1.set_ylabel('Hour')
ax1.set_xlabel('Month')
ax1.plot(df['Day Type'],df['Hour'], c='Equipment Electric Power [kWh]')
ax2.set_ylabel('Hour')
ax2.set_xlabel('Day Type')
plt.show()
plt.close()
```
